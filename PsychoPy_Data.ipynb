## Classification

```python
df = pd.read_csv("G://O meu disco//PhD//1st Study//Results//dataframe_ratings.csv",sep=";",index_col=False)
df
```

```python
columns = df.columns[:10]
columns
```

```python
X = np.array(df[columns])
Y = np.array(df[["Category"]])
```

## Feature Selection

```python
from sklearn import feature_selection, svm
KBest = feature_selection.SelectKBest(feature_selection.f_classif,k=7).fit(X,Y.ravel())
```

```python
features_selected = KBest.get_support()
features_selected
```

Features Removed

```python
columns[features_selected == False]
```

Features Selected

```python
columns[features_selected == True]
```

```python
X_new = X[:,features_selected == True]
```

## Nearest Neighbor

```python
from sklearn import neighbors
n_neighbors=5
```

Shuffle Split

```python
shuffle = sklearn.model_selection.ShuffleSplit(n_splits=10,test_size=0.25)
for train_index,test_index in shuffle.split(X_new):
    X_train, X_test = X_new[train_index],X_new[test_index]
    Y_train, Y_test = Y[train_index],Y[test_index]
```

70% Train - 30% Test

```python
X_train,X_test,Y_train,Y_test = sklearn.model_selection.train_test_split(X_new,Y,test_size=0.3)
```

Classifier

```python
clf = sklearn.neighbors.KNeighborsClassifier(n_neighbors,weights="uniform")
```

```python
clf.fit(X_train,Y_train.ravel())
```

```python
# with open("G://O meu disco//PhD//1st Study//Physiological_Data//Results//knearest.pkl",'rb') as f:
#     clf = pickle.load(f)
# clf
```

```python
Y_pred = clf.predict(X_test)
Y_pred
```

Accuracy Score

```python
print("Accuracy = ", sklearn.metrics.accuracy_score(Y_test,Y_pred)*100)
```

```python
with open("G://O meu disco//PhD//1st Study//Physiological_Data//Results//knearest.pkl",'wb') as f:
    pickle.dump(clf,f)
```
