## Feature Selection

PCA and ICA

```python
algorithms=[]

algorithms.append(("PCA",PCA(n_components=5)))
algorithms.append(("ICA",FastICA(n_components=5)))
```

```python
%matplotlib
for name,algorithm in algorithms:
    X_new = algorithm.fit_transform(X,Y.ravel())
    df_comp = pd.DataFrame(algorithm.components_,columns=columns)
    plt.figure()
    plt.title(name)
    sns.heatmap(df_comp,cmap="plasma")
```

SelectKBest and SelectPercentile

```python
selection=[]
selection.append(("KBest_chi",feature_selection.SelectKBest(feature_selection.chi2,k=5)))
selection.append(("KBest_anova",feature_selection.SelectKBest(feature_selection.f_classif,k=5)))
selection.append(("Percentile_chi",feature_selection.SelectPercentile(feature_selection.chi2,percentile=50)))
selection.append(("Percentile_anova",feature_selection.SelectPercentile(feature_selection.f_classif,percentile=50)))
```

```python
for name,test in selection:
    test.fit(X,Y.ravel())
    print(name)
    print(columns[test.get_support()==False])
```

Reduce Features

```python
KBest = feature_selection.SelectKBest(feature_selection.f_classif,k=4)
```

```python
X_new = KBest.fit_transform(X,Y.ravel())
```

```python
columns[KBest.get_support()==False]
```

Models

```python
models=[]
models.append(("KNN",neighbors.KNeighborsClassifier(n_neighbors=5)))
models.append(("SVM",svm.SVC(C=10,kernel="rbf",random_state=42)))
models.append(("Tree",tree.DecisionTreeClassifier(criterion="gini",random_state=42)))
models.append(("Forest",ensemble.RandomForestClassifier(n_estimators=10)))
models.append(("NN",MLPClassifier(max_iter=100)))
```

```python
cross_validation=[]
cross_validation.append(("KFold",KFold(n_splits=4)))
cross_validation.append(("StratifiedKFold",StratifiedKFold(n_splits=4)))
cross_validation.append(("RepeatedKFold",RepeatedKFold(n_splits=4,n_repeats=2)))
cross_validation.append(("LOO",LeaveOneOut()))
cross_validation.append(("LPO",LeavePOut(2)))
cross_validation.append(("Shuffle",ShuffleSplit(n_splits=4,test_size=0.25)))
```

```python
results=dict()
for name,model in models:
    results_model=dict()
    for validator_name,validator in cross_validation:
        X_train, X_test, Y_train, Y_test = train_test_split(X_new, Y.ravel(), test_size=0.25, random_state=42)
        cv_results = cross_val_score(model,X_train,Y_train.ravel(),cv=validator,scoring="accuracy")
        results_model[validator] = (cv_results.mean(),cv_results.std())
    results[name]=results_model
```

```python
for model in results:
    print(model,"\n")
    for keys,values in results[model].items():
        print(keys,"\n","Score = ",values[0]*100,"\n","STD = ",values[1]*100,"\n")
```
